```python
# Generate some example data
np.random.seed(0)
X = np.random.rand(100, 1)  # Predictor variable
y = 2 * X + 1 + 0.1 * np.random.randn(100, 1)  # Response variable with noise

# Create a DataFrame
data = pd.DataFrame({'X': X.flatten(), 'y': y.flatten()})

# Visualize the data
sns.scatterplot(x='X', y='y', data=data)
plt.title('Scatter plot of the data')
plt.show()

# Add a constant term for the intercept
X_with_intercept = sm.add_constant(X)

# Fit the OLS regression model
model = sm.OLS(y, X_with_intercept)
results = model.fit()

# Display the regression results
print(results.summary())

# Visualize the regression line
sns.scatterplot(x='X', y='y', data=data)
plt.plot(X, results.predict(), color='red', linewidth=2, label='Regression Line')
plt.title('OLS Regression Line')
plt.legend()
plt.show()
```

```python
import pandas as pd

# Load the dataset
file_path = '/content/CLT_Full_Analysis.csv'
df2 = pd.read_csv(file_path)

# Check for missing or infinite values
missing_rows = df2.isnull().any(axis=1)
infinite_rows = df2.isinf().any(axis=1)

# Remove rows with missing or infinite values
df2 = df2[~missing_rows & ~infinite_rows]

# Fit logistic regression model
logit_model = sm.Logit(df2['is_open'], df2[['SentimentQDAP', 'review_count', 'topic_lda', 'rating']])
result = logit_model.fit(disp=0)

# Display summary of the regression
print(result.summary())
```

```python
import pandas as pd
import statsmodels.api as sm

# Load the dataset
file_path = '/content/CLT_Full_Analysis.csv'
df2 = pd.read_csv(file_path)

# Set the dependent variable (outcome) and independent variables
y = df2['is_open']
X = df2[['SentimentQDAP', 'review_count', 'topic_lda', 'rating']]

# Reset indices to ensure proper alignment
y = y.reset_index(drop=True)
X = X.reset_index(drop=True)

# Add a constant term to the independent variables
X = sm.add_constant(X)

# Fit logistic regression model
logit_model = sm.Logit(y, X)
result = logit_model.fit(disp=0)

# Display summary of the regression
print(result.summary())
```
